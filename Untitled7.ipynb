```python
import pandas as pd
import zipfile
import os

# Define file paths
zip_path = "/content/Sector Data zip.zip"  # Replace with your actual zip file
main_df_path = "/content/market-capitalization.csv"  # Replace with your main dataset

# Extract ZIP file
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("extracted_data")  # Extract CSV files into a folder

# Load the main dataframe
main_df = pd.read_csv(main_df_path)

# Get all CSV files from the extracted folder
csv_files = [f for f in os.listdir("extracted_data") if f.endswith(".csv")]

# Merge sector columns
for file in csv_files:
    df = pd.read_csv(f"extracted_data/{file}")

    # Assuming primary key is 'Symbol' and sector column is named 'Sector'
    main_df = main_df.merge(df[['Symbol', 'Sector']], on='Symbol', how='left')

# Save or display the final merged dataframe
main_df.to_csv("merged_data.csv", index=False)
print("Merging complete. Saved as merged_data.csv")
```
